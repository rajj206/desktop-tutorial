```python
#sort the data on unit cost

products_sorted = products.sort_values("Unit Cost", ascending=True) #ascending=False for descending order
products_sorted
```

```python
# sort items in ascending and units in descending order

products_sorted_multi = products.sort_values(["Item", "Units"], ascending=[1,0])
```

```python
# select region and rep columns

products_select = products[["Region", "Rep"]]
products_select
```

```python
#filter the records with unit cost which is greater than 100

products_filter = products[products["Unit Cost"] > 100]
products_filter
```

```python
#filter the records with Units greater than 1 and region as Central
products_filter_1 = products[(products["Units"] > 1) & (products["Region"] == "Central")]
products_filter_1
```

```python
# Filter regions from df which are present in a list

list = ["Central", "East"]
products_filter_2 = products[products["Region"].isin(list)]
products_filter_2
```

```python
# create a new column total cost as a product of Units and Unti cost

products["Total Cost"] = products["Units"] * products["Unit Cost"]
products 
```

```python
#find the mean of unit cost grouping on product

products_mean = products.groupby("Item")["Unit Cost"].mean()
products_mean
```

```python
#find the mean, median , min , max of unit cost grouping on product

products_summary = products.groupby("Item")["Unit Cost"].agg([np.min, np.max, np.mean, np.median])
products_summary
```

```python
# find cumulative sum of total Cost

products_cumulative = products["Total Cost"].cumsum()
products_cumulative
```

```python
# capture disticnt combinations of region and Items 

products_distinct = products.drop_duplicates(subset=["Item", "Region"])
products_distinct[["Region", "Item"]]
```

```python
# count items, different ways

products_count = products["Item"].value_counts()
products_count = products["Item"].value_counts(sort = True, normalize=True) #gives the proportion
products_count
```

```python
# Sum of total Units for Item Pen grouping across regions

products_sum = products[products["Item"] == "Pen"].groupby("Region")["Units"].sum()
products_sum
```

```python
# simple pivot table summary

print (products.pivot_table(index="Region", values="Units", columns="Item", aggfunc=["mean"], fill_value=0))
```
